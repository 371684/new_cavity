```python
import numpy as np
from scipy import interpolate
import pylab as pl
import csv
from pandas import *
import matplotlib.pyplot as plt
```

```python
# reading CSV file

power_scan_data2 = read_csv("2023_4_9_NLTL_powerscan_25MHzstep2.csv")
power_scan_data3 = read_csv("2023_4_9_NLTL_powerscan_25MHzstep3.csv")
power_scan_data4 = read_csv("2023_4_9_NLTL_powerscan_25MHzstep4.csv")
rf_fre2 = power_scan_data2["frequency [Hz]"].tolist()
opt_power2 = power_scan_data2["input_power [dBm]"].tolist()
opt_power3 = power_scan_data3["input_power [dBm]"].tolist()
opt_power4 = power_scan_data4["input_power [dBm]"].tolist()
plt.plot(rf_fre2,opt_power2)
plt.plot(rf_fre2,opt_power3)
plt.plot(rf_fre2,opt_power4)
plt.xlabel("rf frequency/Hz")
plt.ylabel("optimal power/dBm")
plt.title("optimal input power for different frequency")
plt.legend(["scan1","scan2","scan3"])
```

```python
mean_opt_power =np.add(np.add(array(opt_power2), array(opt_power3)),array(opt_power4))/3
plt.plot(rf_fre2, mean_opt_power)
plt.xlabel("rf frequency/Hz")
plt.ylabel("optimal power/dBm")
plt.title("averaged optimal input power")
```

```python
f=interpolate.interp1d(rf_fre2,mean_opt_power,"cubic")
xnew=np.linspace(0.625e9,1.625e9,101)
ynew = f(xnew)
plt.plot(xnew,ynew, label = "quadratic")
plt.xlabel("rf frequency/Hz")
plt.ylabel("optimal power/dBm")
plt.title("interpolation")
```
