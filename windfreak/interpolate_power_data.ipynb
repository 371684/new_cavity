```python
import numpy as np
from scipy import interpolate
import pylab as pl
import csv
from pandas import *
import matplotlib.pyplot as plt
```

```python
# reading CSV file

power_scan_data2 = read_csv("2023_4_9_NLTL_powerscan_25MHzstep2.csv")
rf_fre2 = power_scan_data2["frequency [Hz]"].tolist()
opt_power2 = power_scan_data2["input_power[dBm]"].tolist()
plt.plot(rf_fre2,opt_power2)
```

```python
f=interpolate.interp1d(rf_fre2,opt_power2,"cubic")
xnew=np.linspace(0.625e9,1.625e9,101)
ynew = f(xnew)
plt.plot(xnew,ynew, label = "quadratic")
```
